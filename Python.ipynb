The classification or categorising of data elements is known as data types. It represents the type of value that specifies the operations that can be carried out on a given set of data. In Python programming, everything is an object, hence data types are actually classes and variables are instance (object) of these classes.

Standard Data Types in Python
  1. Numeric
  2. Boolean
  3. Sequence
  4. Set
  5. Dictionary

**Another way of grouping Data types in python can be "primitive" and "non-primitive"**

1. Primitive data types: Also known as "built-in" or "basic" data types, primitive data types are the most fundamental data types in Python. They are predefined by the Python programming language and are used to represent simple values. Some examples of primitive data types in Python include:
    - Integer (int): Represents whole numbers, such as 1, -10, or 1000.
    - Floating-point (float): Represents numbers with decimal points, such as 3.14, -0.5, or 2.0.
    - Boolean (bool): Represents a binary value that can be either True or False.
    - String (str): Represents a string of characters, such as 'a', 'B', or 'abc' etc..

Primitive data types are considered basic building blocks for more complex data structures in Python.

2. Non-primitive data types: Also known as "composite" or "user-defined" data types, non-primitive data types are created by the programmer using one or more primitive data types. They are more complex data structures that can hold multiple values or objects. Some examples of non-primitive data types in Python include:
    - List: Represents a collection of values, where each value can be of any data type. Lists are mutable, meaning their elements can be changed.
    - Tuple: Represents an immutable collection of values, where each value can be of any data type. Tuples cannot be changed after they are created.
    - Set: Represents an unordered collection of unique values, where each value can be of any data type. Sets do not allow duplicate values.
    - Dictionary: Represents a collection of key-value pairs, where each key-value pair is separated by a colon (":") and enclosed in curly braces ("{}"). Dictionaries are used to store and retrieve data using keys rather than indices.

Non-primitive data types are more complex and provide additional functionality beyond basic data types, allowing programmers to create more sophisticated data structures and manipulate data in more advanced ways.

## 1. Numeric

Numeric data type represents the data with numeric value. Integer, floating-point, or even complex number can be seen as numerical value. Python defines these values as belonging to the int, float, and complex classes.

  - Integers - The int class is used to represent this value. There are both positive and negative whole numbers in it (without fraction or decimal). An integer value's length is unrestricted in Python.
  - Float: The float class is used to represent this value. It is a real number that is represented in floating point. There is a decimal point to indicate it. To designate scientific notation, the letters e or E may be added after a positive or negative number.
  - Complex Numbers - The complex class is a representation of a complex number. (Real part) plus (Imaginary part)j is how it is specified. For instance, 4+2j

`NOTE: ` We can use `type()` function to check the datatype.

```python
x = 10
print("Type of x: ", type(x))

y = 15.6
print("\nType of y: ", type(y))

z = 1 + 2j
print("\nType of z: ", type(z))
```

```python

# another method to create a complex number is using the complex() class
a = complex(5, 7)
print("\nType of z: ",a)
```

#### Accessing elements of a string

In python, we can use indexing to access individual elements of a string. Since strings are ordered sequences of characters. Each character has a uniuqe index which tell its positon in the sequence

There are two types of indexing allowed in python.
  1. Positve - It starts from 0 upto length of sequence -1. Here, the first element is indexed 0 and the last index n-1 (n being the length of sequence)
  2. Negative - It starts from -1 upto the -(length of sequence). Here the last element is indexed as -1 and the first element as -(length of sequence)

Example:

       -6 -5 -4 -3 -2 -1  NEGATIVE INDEXING
        ----------------
        p  y  t  h  o  n
        ----------------
        0  1  2  3  4  5  POSITIVE INDEXING

To acess element in string we need to pass the index number as follows

        string[ INDEX VALUE ]

## Type casting

Even though python is a dynamically-typed language there are times when we want to specify a type in to a variable. In python this can be done with casting.

We can use the below functions
- int() - It constructs an integer number from an interger literal, a float literal, or even a string literal provided the string represents a whole number
- float() - It constructs a float number from an interger literal, a float literal, or even a string literal provided the string represents a float or an integer
- str() - It constructs a string from an interger literal, a float literal, or a string literal

```python
marks = input('enter your marks:')
```

```python
type(marks)
```

```python
x = int(marks)
```

```python
type(x)
```

```python
x+3
```

```python
# using int()
x = int(1)
y = int(2.5)
z = int("55")

print("Value of x: ", x, "Data Type of x: ", type(x))
print("Value of y: ", y, "Data Type of y: ", type(y))
print("Value of z: ", z, "Data Type of z: ", type(z))
```

```python
# using float()
x = float(1)
y = float(2.5)
z = float("55")
zz = float("5.5")
print("Value of x: ", x, "Data Type of x: ", type(x))
print("Value of y: ", y, "Data Type of y: ", type(y))
print("Value of z: ", z, "Data Type of z: ", type(z))
print("Value of zz: ", zz, "Data Type of zz: ", type(zz))
```

```python
# using str()
x = str(1)
y = str(2.5)
z = str("55")
zz = str("5.5")

print("Value of x: ", x, "Data Type of x: ", type(x))
print("Value of y: ", y, "Data Type of y: ", type(y))
print("Value of z: ", z, "Data Type of z: ", type(z))
print("Value of zz: ", zz, "Data Type of zz: ", type(zz))
```
